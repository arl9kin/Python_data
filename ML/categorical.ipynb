## Импортируем нужные библиотеки

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Загружаем сырые данные о средней оценке учеников (GPA), результатах стандартизированного теста (SAT) и о посещаемости (если больше 75% занятий посещено, то - Yes)

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
raw_data
```

### Трансформируем качественную переменную в количесвтенную,
где Да = 1, нет = 0

```python
 data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
```

```python
data
```

```python
data.describe()
```

## Создаем регрессию

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

### GPA = 0.6439 + 0.0014 * SAT + 0.2226 * Attendance,
и если Attendance = 0, то уравнение будет GPA = 0.6439 + 0.0014 * SAT
а для Attendance = 1: GPA = 0.8665 + 0.0014 * SAT

```python
sns.set_style('white')
plt.figure (figsize = (8,6))
yhat_no =  0.6439 + 0.0014 * data['SAT']
yhat_yes =  0.8665 + 0.0014 * data['SAT']
plt.scatter(data['SAT'],
            y,
            c = data['Attendance'],
            cmap ='RdYlGn_r',
            s = 80,
            alpha = 0.8)
fig = plt.plot (data['SAT'],
                yhat_no,
                lw=2,
                c = 'red')
fig = plt.plot (data['SAT'],
                yhat_yes,
                lw=3,
                c = 'green')
plt.xlabel ('SAT',
            fontsize = 15,
            fontweight = 'bold')
plt.ylabel ('GPA',
            fontsize = 15,
            fontweight = 'bold')

sns.despine()
plt.show()
```
