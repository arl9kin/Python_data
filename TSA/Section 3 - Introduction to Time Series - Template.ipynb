### Импортируем библиотеки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

### Загружаем данные

```python
raw_data = pd.read_csv('Index2018.csv')
```

```python
df_comp = raw_data.copy()
```

### Исследование данных

### Plotting the Data

```python
plt.figure(figsize = (20,5))
df_comp.spx.plot(title = 'S&P500 prices')
df_comp.ftse.plot(title = 'ftse100 prices')
plt.title ('S&P500 vs ftse100')
plt.show()
```
