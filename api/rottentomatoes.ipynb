```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = 'https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/'
```

```python
response = requests.get(base_site)
response
```

```python
html = response.content
```

```python
html
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
with open ('rotten_tomatoes_lxml.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
divs = soup.find_all('div',{'class':'col-sm-18 col-full-xs countdown-item-content'})
divs
```

```python
headings = [div.find('h2') for div in divs]
headings
```

```python
headings[0]
```

```python
movie_names = [heading.find('a').string for heading in headings]
movie_names
```

```python
movie_years = [heading.find('span', class_= 'start-year').string for heading in headings]
movie_years
```

```python
movie_years[0][1:-1]
```

```python
movie_years[0].strip('()')
```

```python
years = [int(year.strip('()')) for year in movie_years]
years
```

```python
score = [heading.find('span', class_= 'tMeterScore').string for heading in headings]
score
```

```python
consensus = [div.find('div', {'class':"info critics-consensus"}) for div in divs]
consensus
```

```python
common_phrase = 'Critics Consensus: '
```

```python
common_len = len(common_phrase)
```

```python
consensus_text = [con.text[common_len:] if con.text.startswith(common_phrase) else con.text for con in consensus]
consensus_text
```
