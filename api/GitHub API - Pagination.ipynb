# Нумерация страниц

```python
import requests
import json
```

Документация: https://jobs.github.com/api

```python
base_site = "https://jobs.github.com/positions.json"
```

```python
r = requests.get(base_site, params={'location':'Los Angeles'})
r.ok
```

```python
r = r.json()
print(json.dumps(r,indent=4))
```

```python
len(r)
```

### Параметр страница

```python
r = requests.get(base_site).json()
```

```python
len(r)
```

```python
r = requests.get(base_site,params={'page':2}).json()
print(json.dumps(r, indent=4))
```

```python
len(r)
```

```python
r = requests.get(base_site,params={'page':10}).json()
r
```

### Извлечение множества страниц

```python
results = []
```

```python
for i in range(5):
    r = requests.get(base_site,params={'page':i+1}).json()
    
    if len(r) == 0:
        break
    else:
        results.extend(r)
```

```python
len(results)
```
