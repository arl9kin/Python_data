### Создаем Random Walk


```python
step_option = [1,-1]
step_choice = choice(step_option)
# выбираем шаг
```

```python
walk = []
walk.append(step_choice)
for step in range(1,1000):
    next_step = walk[step-1] + choice(step_option)
    walk.append(next_step)
print(walk)    
```

```python
plt.plot(walk)
plt.show()
```

### Используем в виде функции

```python
def rand_walk(step_num):
    walk = []
    step_choice = choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice_2 = choice([1,-1])
        next_step = walk[i-1] + step_choice_2
        walk.append(next_step)
    return walk    
```

```python
random_w = rand_walk(100)
print(random_w)
```

```python
def plot_walk(walk):
    plt.plot(walk)
    plt.xlabel('Кол-во шагов', fontsize = 13, fontweight = 'bold')
    plt.ylabel('Дистанция от начала', fontsize = 13, fontweight = 'bold')
    plt.title('Наш Random Walk', fontsize = 15, fontweight = 'bold')
    plt.show()
```

```python
plot_walk(rand_walk(1000))
```

### Создадим функцию, отображающую несколько random walk.

```python
def plot_walks(num,walk_steps):
    
    lab_list = list(range(1,num+1))
    for i in range(0,num):
        
        x = list(range(1,walk_steps+1))
    
        plt.plot(x,rand_walk(walk_steps),label = 'График номер' + ' ' + str(lab_list[i]))
        plt.xlabel('Кол-во шагов', fontsize = 13, fontweight = 'bold')
        plt.ylabel('Дистанция от начала', fontsize = 13, fontweight = 'bold')
        plt.title('Наш Random Walk', fontsize = 15, fontweight = 'bold')
        plt.legend(loc='best')
    plt.show()
```

```python
plot_walks(5,100000)
```
