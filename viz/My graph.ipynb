```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/stacked_area_chart_data.csv')
```

```python
df
```

```python
sns.set_style ('ticks')
colors = ['black', 'green', 'yellow', 'red']
lebel = ['Diesel','Petrol','Gas','Other']
plt.figure(figsize = (12,9))
plt.stackplot(df['Year'],
              df['Diesel'],
              df['Petrol'],
              df['Gas'],
              df['Other'],
             colors = colors,
             edgecolor = 'none')
plt.title ('My graph', fontsize = 20, fontweight = 'bold')
plt.xticks(df['Year'], rotation = 45)
plt.ylabel ('# of cars', fontsize = 15, fontweight = 'bold')
plt.legend (lebel, loc = 'upper left')
sns.despine()
plt.show()
```
