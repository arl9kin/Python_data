```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/histogram_data.csv')
df
```

```python
sns.set_style ('white')
plt.figure (figsize = (8,6))
plt.hist (df['Price'], color = '#108A99', bins = 8)
plt.title ('Distribution of Real Estate prices', fontsize = 15, fontweight = 'bold', color = 'black', loc = 'left')
plt.ylabel ('Number of Properties', fontsize = 13, fontweight = 'bold', color = 'black')
plt.xlabel ('Price in (000\' $)', fontsize = 13, fontweight = 'bold', color = 'black')
sns.despine()
plt.show()
```
