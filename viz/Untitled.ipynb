```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/bar_line_chart_data.csv')
df
```

```python
sns.set_style('white')
fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(df['Year'],
      df['Participants'],
      color='midnightblue')
ax.set_ylabel('# of Participants',
              fontsize=15,
              weight='bold')
ax.tick_params(axis='y',
               width=1,
               labelsize='large')
ax.tick_params(axis='x',
               width=2,
               labelsize='large')

ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(df['Year'],
         df['Python Users'],
         color='#b60000',
         marker='D')
ax1.set_ylabel('# of Python Users',
               fontsize=15,
               weight='bold',
               color='#b60000')
ax1.tick_params(axis='y',
                width=1,
                labelsize='large',
                colors='#b60000')

ax.set_title('KD Nuggets Survey Python Users (2012-2019)',
            fontsize=18,
            weight='bold',
            loc='left')

sns.despine(right=False,
           bottom=True)
```
