```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/scatter_plot_ii.csv')
df
```

```python
sns.set_style ('white')
plt.figure (figsize =(10,8))

sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df,
           color = 'grey',
           line_kws = {'color':'red'})
plt.xlabel("Ad exp", fontsize = 13, fontweight = 'bold')
plt.ylabel("Sales", fontsize = 13, fontweight = 'bold')
plt.title ("Effect of Ad on Sales", fontsize = 17, fontweight = 'bold', loc = 'left')
sns.despine()
plt.show()
```

```python
sns.set_style ('white')

sns.lmplot(x = 'Budget',
           y = 'Sales',
           data = df,
           scatter_kws = {'color':'k'},
           line_kws = {'color':'red'},
           height = 10)
plt.xlabel("Ad exp", fontsize = 15, fontweight = 'bold')
plt.ylabel("Sales", fontsize = 15, fontweight = 'bold')
plt.title ("Effect of Ad on Sales", fontsize = 20, fontweight = 'bold', loc = 'left')
sns.despine()
plt.show()
```
