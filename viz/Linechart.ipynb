```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/line_chart_data.csv')
```

```python
df.head()
```

```python
df['New_date'] = pd.to_datetime(df['Date'])
df
```

```python
sns.set_style('dark')
lebels = ['S&P500','FTSE100']
plt.figure (figsize = (20,8))
plt.plot(df['New_date'],df['GSPC500'])
plt.plot(df['New_date'], df['FTSE100'])
plt.title('Kek', fontsize = 20, fontweight = 'bold')
plt.ylabel ('returns', fontsize = 15)
plt.xlabel ('Date')
plt.legend (lebels, loc = 'upper left', fontsize = 'large')
plt.show()
```

```python
df_H2_2008 = df[(df.New_date >= '2008-07-01') & (df.New_date <= '2008-12-31')]
df_H2_2008.tail()
```

```python
sns.set_style('dark')
lebels = ['S&P500','FTSE100']
plt.figure (figsize = (20,8))
plt.plot(df_H2_2008['New_date'],df_H2_2008['GSPC500'], color = 'midnightblue')
plt.plot(df_H2_2008['New_date'], df_H2_2008['FTSE100'], color = 'crimson')
plt.title('Kek 2H 2008', fontsize = 20, fontweight = 'bold')
plt.ylabel ('returns', fontsize = 15)
plt.xlabel ('Date')
plt.legend (lebels, loc = 'upper left', fontsize = 'large')
plt.show()
```
