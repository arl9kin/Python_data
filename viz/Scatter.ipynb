```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
```

```python
df = pd.read_csv('C:/study/data/365data/tasks/viz/scatter_data.csv')
df
```

```python
sns.set_style ('white')
plt.figure(figsize = (12,8))
scatter = plt.scatter(df['Area (ft.)'],
            df['Price'],
            c = df['Building Type'],
            alpha = 0.7,
            cmap = 'viridis')
plt.title ('Relationship between Area and Price of California Real Esatate',
           fontsize = 17,
           fontweight = 'bold',
           loc = 'left',
           color = 'midnightblue')
plt.ylabel('Price in (000\'s $)', fontsize = 15, fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontsize = 15, fontweight = 'bold')
plt.legend(*scatter.legend_elements(),
          loc = 'upper left',
          title = 'Building Type')
plt.show()
```

```python
sns.set_style ('darkgrid')
plt.figure(figsize = (12,8))
sns.scatterplot(df['Area (ft.)'],
            df['Price'],
            hue = df['Building Type'],
            alpha = 0.5,
            palette = ['black', 'darkblue', 'purple', 'pink', 'orange'],
            s = 100)

plt.title ('Relationship between Area and Price of California Real Esatate',
           fontsize = 17,
           fontweight = 'bold',
           loc = 'left',
           color = 'midnightblue')
plt.ylabel('Price in (000\'s $)', fontsize = 15, fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontsize = 15, fontweight = 'bold')
plt.legend(*scatter.legend_elements(),
          loc = 'upper left',
          title = 'Building Type')
plt.show()
```
