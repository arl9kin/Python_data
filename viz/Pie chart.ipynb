```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_engine_type = pd.read_csv ('C:/study/data/365data/tasks/viz/pie_chart_data.csv')
```

```python
df_engine_type
```

```python
sns.set_palette ('colorblind')
```

```python
plt.figure (figsize = (10,8))
plt.pie (df_engine_type['Number of Cars'],
        labels = df_engine_type['Engine Fuel Type'].values,
        autopct = '%.1f%%',
        textprops = {'size': 'x-large',
                    'fontweight': 'bold',
                    'rotation': '45',
                    'color': 'w'})
plt.legend()
plt.title ('Kek', fontsize = 18, fontweight = 'bold')
plt.show()
```
